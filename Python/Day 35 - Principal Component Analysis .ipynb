# Principal Component Analysis

PCA is an **Unsupervised** statistical technique used to examine the interrelations among a set of variables and to identigy the underlying structure of the variables. Principal Component Analysis is also referred as a general **factor analysis**.
 - PCA Algorithm is a Technique for transformation of data and attempts to find out what features explain the most variance in the data. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Dataset

We use the bulit-in dataset from Sklearn module which is the cancer data set,since it had so many features/columns.

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
cancer['target_names']

# And our target is 0 and 1.
```

 - In a dataset like above with many features/columns, its not possible to visualize all in a 2d diagram. So, we use PCA to find the first 2 principal components in a 2d diagram.  

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

 - We scale our data so that each variable has a single unit variance to proceed with the PCA.
 - We represent a PCA object, find the principal components using the fit method, then apply the rotation and    dimensionality reduction by calling transform().

 - We can also specify how many components we want to keep when creating the PCA object.

```python
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)

# We ultimately try to visualize the 30 dimensional dataset in 2 principal Components 
```

```python
pca.fit(scaled_data)
```

Now we can transform this data to its first 2 principal components.

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

Note: We reduced the scaled data from 30 dimensions to 2 PC's 

```python
plt.figure(figsize=(8,6),dpi=100)
plt.scatter(x_pca[:,0],x_pca[:,1])
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

 - From the above plot, we successfully plotted the first and second Principal components, but we are not able to classify them!

```python
plt.figure(figsize=(8,6), dpi=100)
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='Set1')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

 - Above, We have just passed a color parameter based on the target variable. We can now see a clear demarcation between the 2 classes

## Interpreting the components 

Unfortunately, dimensionality reduction can be powerful but is not easy understand what these components represent.

The components correspond to combinations of the original features, the components themselves are stored as an attribute of the PCA object:

```python
pca.components_
```

 - In the above matrix array, each row represents a principal component, and each column relates back to the original features.
 - We can use a heatmap to visualize on what these principal Components represent.

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_comp

# We can see the 1st and 2nd Principal Components and their relationship with the other 30 dimensions as a dataframe.
```

```python
plt.figure(figsize=(12,6),dpi=100)
sns.heatmap(df_comp,cmap='coolwarm')
```

This heatmap and the color bar represents the correlation between the various feature and the principal components.
