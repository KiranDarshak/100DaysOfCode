# Distribution Plots

The plots that allow us to visualize the distribution of a data set are:

* distplot
* jointplot
* pairplot
* rugplot

```python
import seaborn as sns
%matplotlib inline
```

 - We are using the built-in dataset available in Seaborn

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

## 1. distplot

The distplot shows the distribution of a univariate set of observations.

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)

# Note the KDE (Kernel Density Estimation) is missing
```

## 2.  jointplot

jointplot( ) allows to match up two distplots for bivariate data. The parameter to compare with: 
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

## 3. pairplot

Pairplot represents pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```
