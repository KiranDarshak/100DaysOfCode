# Regression Plots

Seaborn has many built-in capabilities for regression plots. **lmplot** allows you to display linear models, but it also allows us to split up those plots based off of features.

```python
import seaborn as sns
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## lmplot( )

```python
sns.lmplot(x='total_bill',y='tip',data=tips, height=6)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', height=7)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm', height=7)
```

### Working with Markers

lmplot kwargs get passed through to **regplot** which is a more general form of lmplot(). regplot has a scatter_kws parameter that gets passed to plt.scatter. So, set the s parameter in that dictionary, which corresponds to the squared markersize.In this case, s for size of a scatter plot.


```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['+','*'],scatter_kws={'s':50},height=8)
```

## Using a Grid

We can add more variable separation through columns and rows with the use of a grid. Just indicate this with the col or row arguments:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',height=6)
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips,height=6)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',aspect=0.6,height=6)
```

### Practice with Titanic dataset

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic,color='navy',height=7)
```

```python
fig = plt.figure(figsize=(15,5))
sns.distplot(titanic['fare'],bins=30,kde=False,color='red')
```

```python
fig = plt.figure(figsize=(12,8))
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
fig = plt.figure(figsize=(12,8),dpi=80)
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

```python
fig = plt.figure(figsize=(12,8))
sns.countplot(x='sex',data=titanic)
```

```python
fig = plt.figure(figsize=(12,8),dpi=80)
sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('titanic_Correlation')
```

```python
g = sns.FacetGrid(data=titanic,col='sex',height=6)
g.map(plt.hist,'age')
```
