# Indexing and Slicing
Indexing is used for calling a single Character in a string


            string  = H  e  l  l  o

    Index position  = 0  1  2  3  4

     Reverse Index  = 0 -1 -2 -3 -4
     
# Syntax for Slicing [start:stop:step/jump]

```python
# Dividing a sentence into two parts

statement = ('This is line one \n this is line two')
```

```python
print(statement)
```

```python
statement[2]
# from statement variable i am calling 2 nd index.
```

```python
mystring =('pencil')
```

```python
mystring[-2]
# negative index
```

```python
mystring[3]
```

```python
alpha ='abcdefgh'
```

```python
alpha
```

```python
alpha[2:]
# Slicing from position 2 (C) to the end of string
```

```python
alpha[:5] # upto 5 but not 5th index
# Slicing all Characters Beginning till 4th index (e)
#NOTE - I called for the 5th index but got the 4th index character as result as it starts from 0
```

```python
alpha[1:6]
```

```python
alpha[::]
# Prints all characters of the string
```

```python
alpha[::2]
# There is 2 character skip/jump in the string
```

```python
alpha[1:6:2]
# all characters from 1st to 6th index - b,c,d,e,f
# jump 2 characters as a condition- b,d,f
```

```python
alpha[::-1]
#reverses the entire string
```

# V. String Formatting

Syntax - 'String here {} then also {}'.format('something1','something2')

```python
print('This is a string {}'.format('Formatting '))
```

```python
print ('The {} {} {}'.format ('is','wrong','order'))

# The sentence is meaningless as the words are jumbled.
```

```python
print ('The {1} {0} {2}'.format ('is','order','wrong'))

# We use index positioning here
```

```python
print ('The {o} {i} {w}'.format (i='is',o='order',w='wrong'))

# Use varable assignment for text formatting
```

```python
print('{0:8} | {1:9}'.format('Spending', 'Amount'))
print('{0:8} | {1:9}'.format('Grocery', 80))
print('{0:8} | {1:9}'.format('Rent', 1000))
```

# Float formatting follows "result: {value:{width}.{precision}}"

```python
results = 2000/33
print(results)
```

```python
print ('The result was {r:1.2f}'.format(r=results))
```
