# K Means Clustering Algorithm

K Means Clustering is an **unsupervised** learning algorithm that tries to cluster data based on their similarity. Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data.

In k means clustering, we have the specify the number of clusters we want the data to be grouped into. The algorithm randomly assigns each observation to a cluster, and finds the centroid of each cluster. Then, the algorithm iterates through two steps:
1. Reassign data points to the cluster whose centroid is closest. 
2. Calculate new centroid of each cluster.


These two steps are repeated till the within cluster variation cannot be reduced any further. The within cluster variation is calculated as the sum of the distance between the data points and their respective cluster centroids.

## K Means Cluster

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)

# We specified 2 clusters because we know them beforehand, but in actual dataset there will be no labels.
```

```python
kmeans.fit(df.drop('Private',axis=1))

# We exclude the private label.
```

```python
kmeans.cluster_centers_
```

## Evaluation

There is no way to evaluate clustering if we don't have the labels, however since this is just to learn, we do have the labels, so we take advantage of this to evaluate our clusters, But, we won't have this advantage in the real world.

```python
def converter(cluster):
    if cluster=='Yes':
        return 0
    else:
        return 1
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```

**Not a Bad result,considering that the algorithm is purely using the features to cluster the universities into 2 distinct groups**
