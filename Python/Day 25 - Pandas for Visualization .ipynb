# Pandas for Data Visualization

pandas has built-in capabilities for data visualization and It is developed from matplotlib library.

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

## The Data

There are some data csv files format. We can read them as dataframes:

```python
data1 = pd.read_csv('data1.csv',index_col=0)
data2 = pd.read_csv('data2.csv')
```

## Style Sheets

Matplotlib has different plot styles to make the plots look nicer. These style sheets include plot_bmh,plot_fivethirtyeight,plot_ggplot and more. They basically create a set of style rules that the plots follow. It's better using them, they make all the plots have the same look and appear more professional.


**Before plt.style.use( ) your plots look like this:**

```python
data1['A'].hist(figsize=(12,6))
```

```python
plt.style.use('ggplot')
```

```python
data1['A'].hist(figsize=(12,6))
```

```python
plt.style.use('bmh')
data1['A'].hist(figsize=(12,6))
```

```python
plt.style.use('dark_background')
data1['A'].hist(figsize=(12,6))
```

```python
plt.style.use('fivethirtyeight')
data1['A'].hist(figsize=(12,6),grid=True)
```

```python
plt.style.use('ggplot')
```

# Plot Types

There are several plot types built-in to pandas, most of them statistical plots by nature:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

We can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)


## Area

```python
data2.plot.area(alpha=0.4,figsize=(12,6))
```

## Barplots

```python
data2.plot.bar(figsize=(12,6))
```

```python
data2.plot.bar(stacked=True,figsize=(12,6))
```

## Histograms

```python
data1['A'].plot.hist(bins=50,grid=True,figsize=(12,6))
```

## Scatter Plots

```python
data1.plot.scatter(x='A',y='B',grid=True,figsize=(12,6))
```

```python
data1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm',s=data1*75,figsize=(12,6))
```

```python
data1.plot.scatter(x='A',y='B',s=data1['C']*75,figsize=(12,6))
```

## BoxPlots

```python
data2.plot.box(figsize=(12,6))
```

## Hexagonal Bin Plot

Used for Bivariate Data, alternative to scatterplot

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges',figsize=(12,6))
```

## Kernel Density Estimation plot (KDE)

```python
data2['a'].plot.kde(figsize=(12,6))
```

```python
data2.plot.density(figsize=(12,6))
```
